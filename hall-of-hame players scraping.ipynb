```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

# generate a data of players from the hall of fame member

```python
url = 'https://en.hispanosnba.com/players/hall-of-fame/index'
```

```python
requ = requests.get(url)
soup = BeautifulSoup(requ.content , 'html.parser')
```

```python
#get the url of players
table = soup.find('tbody')
url_start = 'https://en.hispanosnba.com'
find_a = table.find_all('a')

url_pls=[]
for th in find_a:
    th = str(th)
    th = th.split('"')
    url_pls = url_pls +[url_start+th[1]+'/stats']
#now url_pls contain all the url of players
```

```python
#get the players name
player_name = []
for th in find_a:
    th = str(th)
    th = th.split('"')
    player_name = player_name + [th[3]]
#player name contain now the names of all the players
```

```python
#get tea stats of the players
player_stats=[]
for url in url_pls:
    url_a = url
    requ_a = requests.get(url_a)
    soup_a =  BeautifulSoup(requ_a.content , 'html.parser')
    total_table = soup_a.find('tfoot')
    tr = total_table.find('tr')
    tds = tr.find('td')
    tds = str(tds)
    tds = tds.split('<td>')
    tds[14] = tds[14][8:-9]
    tds[15] = tds[15][:-75]
    player_stats = player_stats + [tds]
#player_stats contain now the stat of all the players
```

```python
#generate a empty table
d = {'PLAYER_NAME':[] , 'TEAM':[] , 'GP':[] , 'GS':[] , 'MIN':[] , 'FG':[] , '3P':[] , 'FT':[] ,
     'REB':[] ,'AST':[] , 'STL':[] , 'TO':[] , 'BLK':[] , 'PF':[] , 'PTS':[] , 'EFF':[]}
df = pd.DataFrame(columns = d)


# 1) generate a small table with just one player for every player
# 2) using pd.concat for add the small table to the data frame
i = 0
for stats in player_stats:
    data = {'PLAYER_NAME':[stats[0]], 'TEAM':[stats[1]], 'GP':[stats[2]], 'GS':[stats[3]], 'MIN':[stats[4]], 
            'FG':[stats[5]], '3P':[stats[6]], 'FT':[stats[7]], 'REB':[stats[8]], 'AST':[stats[9]], 'STL':[stats[10]], 
            'TO':[stats[11]], 'BLK':[stats[12]], 'PF':[stats[13]], 'PTS':[stats[14]], 'EFF':[stats[15]]}
    player_row = pd.DataFrame(data , index = [i])
    i = i+1
    df = pd.concat([df , player_row])
```

```python
#add the players name to the data frame
df['PLAYER_NAME'] = player_name

#set the index to "player_name"
df.set_index('PLAYER_NAME', inplace=True)

#add "hall of fame" column to the data frame
df["H.O.F"] = 1
```

# generate a data of players that not considerd as a Hall of Fame member

```python
#generate urls for the url to the stats of the players
abc = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t', 'u' , 'v' , 'w' , 'y' , 'z']
url_abc = []
url = 'https://en.hispanosnba.com/players/nba-all/'
for a in abc:
    url_abc = url_abc + [url + a]
#now url_abc contain all url for list of players in order to a,b,c
```

```python
#get the url of every player
url_players = [] 
url_start = 'https://en.hispanosnba.com'
for url in url_abc:
    requ_b = requests.get(url)
    soup_b = BeautifulSoup(requ_b.content , 'html.parser')
   
    table_b = soup_b.find('tbody')
    find_name = table_b.find_all('a')
    
    for name in find_name:
        name = str(name)
        name = name.split('"')
        url_players = url_players + [url_start + name[1] + '/stats']
#now url_players contain the url of all the players in the NBA 
```

```python
#get the players name
players_name = []
for url_name in url_abc:
    requ_name = requests.get(url_name)
    soup_name = BeautifulSoup(requ_name.content , 'html.parser')
    tbody = soup_name.find('tbody')
    a_name = tbody.find_all('a')
    for name in a_name:
        name = str(name)
        name = name.split('"')
        players_name = players_name + [name[3]]
#now players_name contain the names of all the players
```

```python
#un_stats contain players index without stats
un_stats = []

player_stats_c = []
i = 0
for url_c in url_players:
    try:  
        requ_c = requests.get(url_c)
        soup_c =  BeautifulSoup(requ_c.content , 'html.parser')  
        total_table_c = soup_c.find('tfoot')
        tr_c = total_table_c.find('tr')
        tds_c = tr_c.find('td')
        tds_c = str(tds_c)
        tds_c = tds_c.split('<td>')
        tds_c[14] = tds_c[14][8:-9]
        tds_c[15] = tds_c[15][:-75]
        player_stats_c = player_stats_c + [tds_c]
    except: 
       # print('problem')
       # print(i)
        un_stats = un_stats + [i] 
        player_stats_c = player_stats_c + [None]
    i = i + 1
```

```python
#generate more data
df_allstar = pd.read_html('https://en.hispanosnba.com/nba/history/awards-and-honors/all-star-game-selections')
df_month = pd.read_html('https://en.hispanosnba.com/nba/history/awards-and-honors/player-of-the-month')
df_teams = pd.read_html('https://en.hispanosnba.com/nba/history/awards-and-honors/all-nba-team-selections')
df_defense = pd.read_html('https://en.hispanosnba.com/nba/history/awards-and-honors/all-defensive-team-selections')

#edit the tables
df_allstar = df_allstar[0]
df_month = df_month[0]
df_teams = df_teams[0]
df_defense = df_defense[0]

#drop coulmns
df_allstar.drop('Seasons' , axis=1, inplace=True)
df_month.drop('Seasons' , axis=1, inplace=True)

#edit name of coulmns
df_allstar = df_allstar.rename({'Player':'PLAYER_NAME'}, axis=1) 
df_month = df_month.rename({'Player':'PLAYER_NAME'}, axis=1)
df_team = df_teams.rename({'Player':'PLAYER_NAME'}, axis=1)
df_defense = df_defense.rename({'Player':'PLAYER_NAME'}, axis=1)

#set index of data frams
df_allstar = df_allstar.set_index('PLAYER_NAME') 
df_month = df_month.set_index('PLAYER_NAME')
df_team = df_team.set_index('PLAYER_NAME') 
df_defense = df_defense.set_index('PLAYER_NAME')
```

```python
#get more stats about the players

data_e =[]
Retired = []

for url in url_players:
    try:
        url = url[:-6]
        requ_c = requests.get(url)
        soup_c = BeautifulSoup(requ_c.content , 'html.parser')
        data_c = soup_c.find('section'  , class_='blkjug block quad')
        data_c = str(data_c)
        data_c = data_c.split('>')
       
        #generate check if the player was Retired 
        data_d = soup_c.find('span' , class_='jugequ')
        data_d = str(data_d)
        data_d = data_d.split('s">')
        data_d = data_d[1][:-11]
        
        data_e = data_e + [[data_c[7].split('<')[0][1]] +  [data_c[12].split('<')[0][1]] + 
                           [data_c[17].split('<')[0][1]] + [data_c[22].split('<')[0][1]]]
        Retired = Retired + [data_d]
    except:
        data_e = data_e + [None]
        Retired = Retired + [None]
```

```python
for i in range(len(Retired)):
    if Retired[i] == 'Retired' or Retired[i] == 'Hall of Fame member':
        Retired[i] = 1
    else:
        Retired[i] = 0
```

```python
more_stats_d = {'PLAYER_NAME':[] , 'Seasons':[], 'Playoffs':[] , 'Finals':[] , 'titles':[] , 'Retired':[]}
more_stats = pd.DataFrame(more_stats_d)

un_stats_a = []
i = 0
for data in data_e:
    if data == None or Retired[i] == None:
        more_stats_d = {'PLAYER_NAME':[players_name[i]] , 'Seasons':[None], 'Playoffs':[None] , 'Finals':[None] , 'titles':[None] , 'Retired':[None]}
        un_stats_a = un_stats_a + [i]
    else:
        more_stats_d = {'PLAYER_NAME':[players_name[i]] , 'Seasons':[data[0]], 'Playoffs':[data[1]] , 'Finals':[data[2]] , 'titles':[data[3]] , 'Retired':[Retired[i]]}
    player_row_c = pd.DataFrame(more_stats_d , index = [i])
    more_stats = pd.concat([more_stats , player_row_c])
    i = i + 1
more_stats = more_stats.set_index('PLAYER_NAME')
```
